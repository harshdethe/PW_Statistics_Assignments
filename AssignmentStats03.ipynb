1. What is hypothesis testing in statistics?
   - Hypothesis testing in statistics is a procedure used to make decisions about a population based on sample data. It involves formulating two competing statements: the null hypothesis ($H_0$), which represents no effect or no difference, and the alternative hypothesis ($H_a$), which represents the presence of an effect or difference.The process begins by stating these hypotheses and choosing a significance level (alpha), typically 0.05, which defines the threshold for rejecting the null hypothesis. Data is then collected, and a test statistic is calculated to measure how much the sample data deviates from what is expected under the null hypothesis. Using this test statistic, a p-value is determined, which indicates the probability of observing the data if the null hypothesis is true. If the p-value is less than or equal to the significance level, the null hypothesis is rejected in favor of the alternative hypothesis; otherwise, there is not enough evidence to reject the null hypothesis.

2. What is the null hypothesis, and how does it differ from the alternative
hypothesis?
   - The null hypothesis ($H_0$) is a statement that assumes no effect, no difference, or no relationship exists in the population. It serves as the default or baseline assumption that any observed effect in the sample data is due to random chance. For example, if testing whether a new drug has an effect, the null hypothesis might state that the drug has no effect on patients.The alternative hypothesis ($H_a$ or $H_1$) is the statement that contradicts the null hypothesis. It represents the presence of an effect, difference, or relationship that the researcher aims to support. Continuing the drug example, the alternative hypothesis would state that the drug does have an effect on patients.


3. Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.
   - The significance level, denoted by alpha symbol, is a predetermined threshold in hypothesis testing that represents the maximum acceptable probability of making a Type I error, which occurs when the null hypothesis is wrongly rejected even though it is true. Typically set at 0.05, the significance level indicates a 5% risk of such an error. When conducting a test, the p-value—representing the probability of obtaining the observed data assuming the null hypothesis is true—is compared to $\alpha$. If the p-value is less than or equal to the significance level, the result is considered statistically significant, leading to the rejection of the null hypothesis. Conversely, if the p-value is greater than $\alpha$, there is not enough evidence to reject the null hypothesis.

4. What are Type I and Type II errors? Give examples of each.
   - Type I and Type II errors are two types of mistakes that can occur in hypothesis testing. A Type I error happens when the null hypothesis is true, but it is incorrectly rejected. This means concluding that there is an effect or difference when, in reality, there isn’t one. For example, in a medical trial, a Type I error would occur if a new drug is declared effective when it actually has no real benefit. On the other hand, a Type II error occurs when the null hypothesis is false, but it is mistakenly not rejected. This means failing to detect a real effect or difference. Using the same example, a Type II error would be concluding that the new drug is ineffective when it actually works

5. What is the difference between a Z-test and a T-test? Explain when to use
each.
   - The main difference between a Z-test and a T-test lies in the information available about the population and the sample size. A Z-test is used when the population variance is known and the sample size is large (typically n > 30), allowing the test statistic to follow a standard normal distribution. It is commonly applied to test hypotheses about population means or proportions when these conditions are met. In contrast, a T-test is used when the population variance is unknown and the sample size is small. The T-test relies on the Student’s t-distribution, which accounts for the additional uncertainty from estimating the population variance from the sample. Therefore, the T-test is appropriate for smaller samples or when the population standard deviation is not known.

```python
# 6 Write a Python program to generate a binomial distribution with n=10 and p=0.5, then plot its histogram.
import numpy as np
import matplotlib.pyplot as plt

# Parameters
n = 10      # number of trials
p = 0.5     # probability of success
size = 1000 # number of samples to generate

# Generate binomial distribution samples
data = np.random.binomial(n, p, size)

# Plot histogram
plt.hist(data, bins=range(n+2), align='left', edgecolor='black', density=True)
plt.title('Binomial Distribution (n=10, p=0.5)')
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.xticks(range(n+1))
plt.show()
```

```python
# 7. Implement hypothesis testing using Z-statistics for a sample dataset in
# Python. Show the Python code and interpret the results.
# sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
#                         50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
#                         50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
#                         50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

import numpy as np
from scipy.stats import norm

# Sample data
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

# Known population standard deviation
population_std = 0.5

# Hypothesized population mean
mu_0 = 50

# Sample statistics
sample_mean = np.mean(sample_data)
sample_size = len(sample_data)

# Calculate Z-statistic
z_stat = (sample_mean - mu_0) / (population_std / np.sqrt(sample_size))

# Calculate p-value for two-tailed test
p_value = 2 * (1 - norm.cdf(abs(z_stat)))

print(f"Sample mean: {sample_mean:.3f}")
print(f"Z-statistic: {z_stat:.3f}")
print(f"P-value: {p_value:.4f}")

# Significance level
alpha = 0.05

if p_value <= alpha:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")
```

```python
#8  Write a Python script to simulate data from a normal distribution and
# calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.

import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Parameters for the normal distribution
mu = 0       # true mean
sigma = 1    # true standard deviation
n = 100      # sample size

# Simulate data
data = np.random.normal(mu, sigma, n)

# Calculate sample mean and standard error
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)
se = sample_std / np.sqrt(n)

# Calculate 95% confidence interval using t-distribution
confidence = 0.95
df = n - 1
t_crit = stats.t.ppf((1 + confidence) / 2, df)
margin_of_error = t_crit * se
ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error

print(f"Sample mean: {sample_mean:.4f}")
print(f"95% confidence interval for the mean: ({ci_lower:.4f}, {ci_upper:.4f})")

# Plot the data
plt.figure(figsize=(10, 6))
plt.hist(data, bins=15, alpha=0.7, color='skyblue', edgecolor='black')
plt.axvline(sample_mean, color='red', linestyle='--', label='Sample Mean')
plt.axvline(ci_lower, color='green', linestyle='--', label='95% CI Lower')
plt.axvline(ci_upper, color='green', linestyle='--', label='95% CI Upper')
plt.title('Histogram of Simulated Normal Data with 95% CI for Mean')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
# 9. Write a Python function to calculate the Z-scores from a dataset and
# visualize the standardized data using a histogram. Explain what the Z-scores represent
# in terms of standard deviations from the mean.

import numpy as np
import matplotlib.pyplot as plt

def cal_Zscore(data):
    data = np.array(data)
    mean = np.mean(data)
    std = np.std(data, ddof=1)  # sample standard deviation

    # Calculate Z-scores
    z_scores = (data - mean) / std

    # Plot histogram of Z-scores
    plt.figure(figsize=(8, 5))
    plt.hist(z_scores, bins=15, color='lightcoral', edgecolor='black', alpha=0.7)
    plt.title('Histogram of Z-scores (Standardized Data)')
    plt.xlabel('Z-score')
    plt.ylabel('Frequency')
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.show()

    return z_scores

# Example usage:
data = np.random.normal(10, 2, 100)
z_scores = cal_Zscore(data)
```
